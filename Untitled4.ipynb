```python
# Cell 1 — Install (force CUDA PyTorch before ultralytics)
!pip install torch torchvision --index-url https://download.pytorch.org/whl/cu121 -q
!pip install ultralytics -q

from ultralytics import YOLO
import yaml, os, zipfile
import torch
from google.colab import drive
```

```python
# Cell 2 — Mount Drive, verify GPU, extract dataset
drive.mount('/content/drive')

device = 0 if torch.cuda.is_available() else 'cpu'
print(f"Using device: {'GPU (' + torch.cuda.get_device_name(0) + ')' if device == 0 else 'CPU — stop and enable GPU in Runtime settings'}")

!cp "/content/drive/MyDrive/yolo_training_data.zip" /content/yolo_training_data.zip

with zipfile.ZipFile('/content/yolo_training_data.zip', 'r') as z:
    z.extractall('/content/dataset')

print("Extracted:")
!ls /content/dataset
```

```python
# Cell 3 — Verify dataset structure
DATASET_ROOT = '/content/dataset/yolo_training_data'

for root, dirs, files_list in os.walk(DATASET_ROOT):
    level = root.replace(DATASET_ROOT, '').count(os.sep)
    if level > 2: continue
    print('  ' * level + os.path.basename(root) + '/')

print("\nImage & label counts:")
for split in ['train', 'valid', 'test']:
    img_dir = os.path.join(DATASET_ROOT, split, 'images')
    lbl_dir = os.path.join(DATASET_ROOT, split, 'labels')
    if os.path.exists(img_dir):
        n_img = len([f for f in os.listdir(img_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
        n_lbl = len([f for f in os.listdir(lbl_dir) if f.endswith('.txt')]) if os.path.exists(lbl_dir) else 0
        print(f"  {split}: {n_img} images, {n_lbl} labels")
```

```python
# Cell 4 — Create data.yaml
DATASET_ROOT = '/content/dataset/yolo_training_data'

data = {
    'path': DATASET_ROOT,
    'train': 'train/images',
    'val': 'valid/images',
    'test': 'test/images',
    'nc': 14,
    'names': [
        'floor_drain',
        'drain_outlet',
        'air-gap_fitting',
        'drain_line_vent',
        'bin_drain_line',
        'drain_grate',
        'installation_date',
        'aqua_sticker',
        'qr_code',
        'water_stop',
        'water_line',
        'pressure_gauge',
        'water_filter',
        'data_sticker'
    ]
}

yaml_path = f'{DATASET_ROOT}/data.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(data, f, default_flow_style=False, allow_unicode=True)

print("data.yaml:")
with open(yaml_path) as f:
    print(f.read())
```

```python
# Cell 5 — Train
PROJECT_DIR = '/content/drive/MyDrive/yolo_training'
os.makedirs(PROJECT_DIR, exist_ok=True)

model = YOLO('yolov8l.pt')

results = model.train(
    data=f'{DATASET_ROOT}/data.yaml',
    epochs=300,
    imgsz=640,
    batch=16,        # drop to 8 if out-of-memory
    patience=50,     # early stopping

    # Augmentations matching your Roboflow setup:
    mosaic=1.0,
    degrees=15.0,    # rotation
    scale=0.5,       # crop/scale jitter
    hsv_s=0.7,       # saturation (glare/lighting)
    hsv_v=0.4,       # brightness
    fliplr=0.5,
    translate=0.1,

    project=PROJECT_DIR,
    name='yolov8l_run',
    save=True,
    save_period=50,  # checkpoint every 50 epochs
    device=device
)
```

```python
# Cell 6 — Validate & Export
best = YOLO(f'{PROJECT_DIR}/yolov8l_run/weights/best.pt')
metrics = best.val(data=f'{DATASET_ROOT}/data.yaml')

print(f"mAP50:     {metrics.box.map50:.4f}")
print(f"mAP50-95:  {metrics.box.map:.4f}")
print(f"Precision: {metrics.box.mp:.4f}")
print(f"Recall:    {metrics.box.mr:.4f}")

best.export(format='onnx', imgsz=640)
print(f"\nWeights saved to: {PROJECT_DIR}/yolov8l_run/weights/")
```
